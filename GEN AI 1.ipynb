```python
# Experiment: HMM based Next Word Prediction using Brown Corpus

import nltk
import random
from nltk.corpus import brown
from nltk.tag import hmm

# Step 1: Download dataset
nltk.download('brown')
nltk.download('universal_tagset')

# Step 2: Load tagged sentences
tagged_sentences = brown.tagged_sents(tagset='universal')

# Step 3: Split into train data
train_data = tagged_sentences[:5000]

print("Number of training sentences:", len(train_data))
print("Example sentence:", train_data[0])

# Step 4: Train HMM Tagger
trainer = hmm.HiddenMarkovModelTrainer()
hmm_tagger = trainer.train(train_data)

# Step 5: Function to predict next word
def predict_next_hmm(word):
    word = word.lower()

    try:
        # Get tag of input word
        current_tag = hmm_tagger.tag([word])[0][1]
    except:
        return None

    # Get transition probabilities
    next_tag_dist = hmm_tagger._transitions[current_tag]

    # Convert to list of (tag, prob)
    next_tag_probs = [(tag, next_tag_dist.prob(tag)) for tag in next_tag_dist.samples()]

    if not next_tag_probs:
        return None

    # Choose next tag based on probability
    tags, weights = zip(*next_tag_probs)
    next_tag = random.choices(tags, weights=weights)[0]

    # Get emission probabilities for next tag
    emission_dist = hmm_tagger._outputs[next_tag]

    word_probs = [(w, emission_dist.prob(w)) for w in emission_dist.samples()]

    if not word_probs:
        return None

    # Choose next word
    words, weights = zip(*word_probs)
    next_word = random.choices(words, weights=weights)[0]

    return next_word


# Step 6: Test words
test_words = ["he", "is", "will", "have"]

for w in test_words:
    print(f"{w} -> {predict_next_hmm(w)}")


# Result
print("\nResult: HMM model for next word prediction implemented successfully using Brown corpus.")
```
