```python
# Experiment 2:
# Implementation of Skip-gram model using Word2Vec on Brown Corpus

# Step 1: Import required libraries
import nltk
from nltk.corpus import brown
from gensim.models import Word2Vec
import matplotlib.pyplot as plt

# Step 2: Download and load dataset
nltk.download('brown')

# Load sentences from Brown corpus
sentences = brown.sents()

# Step 3: Select training data
train_sentences = sentences[:5000]

print("No. of training sentences:", len(train_sentences))
print("Example sentence:", train_sentences[0])

# Step 4: Train Skip-gram Word2Vec model
model = Word2Vec(
    sentences=train_sentences,
    vector_size=100,
    window=5,
    min_count=2,
    sg=1 # Skip-gram model
)

# Step 5: Access word embeddings
print("\nEmbedding for 'money':")
print(model.wv['money'])

# Step 6: Find similar words
print("\nWords similar to 'money':")
print(model.wv.most_similar('money', topn=10))

# Step 7: Similarity between two words
print("\nSimilarity between 'man' and 'woman':")
print(model.wv.similarity('man', 'woman'))

# Step 8: Save and load model
model.save("skipgram_model.model")

# Load model
loaded_model = Word2Vec.load("skipgram_model.model")

# Step 9: Test loaded model
print("\nMost similar words to 'good':")
print(loaded_model.wv.most_similar('good'))

# Result
print("\nResult: The Skip-gram model was successfully trained and word embeddings were generated.")
```
